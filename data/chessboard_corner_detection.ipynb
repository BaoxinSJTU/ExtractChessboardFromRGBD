```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
image = cv.imread("color/0.png", cv.IMREAD_COLOR)
```

```python
def imshow(src: np.ndarray):
    im_show = src[:,:,::-1]
    plt.imshow(im_show)
    plt.show()
```

```python
imshow(image)
```

```python
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
_,bin_gray = cv.threshold(gray, 150, 255, cv.THRESH_BINARY)
```

```python
plt.imshow(bin_gray, cmap='gray');
```

```python
retval, corners = cv.findChessboardCorners(image=gray, patternSize=(4,4))
```
